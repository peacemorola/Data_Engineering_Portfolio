```python
#import section
from typing import Dict, List, Tuple, Any
from datetime import datetime, timedelta
from collections import Counter
from tabulate import tabulate

#Book store
book: Dict[str, Dict] = {
    "thermodynamics":{
        'tag': 'AU001',
        'author': 'Robert T Smith',
        'doe': "09-08-25", #date of entry(doe)
        'copies': 2,
        'genre': 'Science',
    },
    
     "automatic control system":{
        'tag': 'AU002',
        'author': 'Kiyosaki J Sheng',
        'doe': "01-01-21",
        'copies': 3,
        'genre': 'Electronics',
    },
     
      "fluid mechanics":{
        'tag': 'AU003',
        'author': 'Meghan R Dumbledor',
        'doe': "06-04-2023",
        'copies': 5,
        'genre': 'Engineering',

    }
}

student: Dict[str, Dict[str, Any]] = {
    "STU001":{
        'name': 'Oyeyemi Morolaoluwagba',
        'department': 'Mechanical engineering',
        'borrowing_details':{
            'borrowed_book': ['fluid mechanics'],
            'borrow_date': [datetime(2025, 9, 9 )],
            'return_date': [datetime(2025, 9, 16)],
            'borrowing_history': ['fluid mechanics', 'automatic control system', 'thermodynamics']

        },
        
    },
    "STU002":{
        'name': 'Yemi Osinbajo',
        'department': 'Mathematics',
        'borrowing_details':{
            'borrowed_book': ['fluid mechanics'],
            'borrow_date': [datetime(2025, 7, 7 )],
            'return_date': [datetime(2025, 9, 14)],
            'borrowing_history': ['fluid mechanics', 'automatic control system', 'thermodynamics']
        },
    },
    "STU003":{
        'name': 'Tolani Badmus',
        'department': 'Political science',
        'borrowing_details':{
            'borrowed_book': ['fluid mechanics'],
            'borrow_date': [datetime(2025, 8,1 )],
            'return_date': [datetime(2025, 8, 18)],
            'borrowing_history': ['fluid mechanics', 'automatic control system', 'thermodynamics']
      }
    }
}
#book checker    
def book_exists(bid:str) -> bool:
    return bid in book
#student checker    
def student_exists(sid:str) -> bool:
    return sid in student
#borrow checker    
def borrowed_book_checker (sid:str, check_history:str):
    if check_history == "yes":
        n_b_b = len(student[sid]["borrowing_details"]["borrowed_book"])
        if n_b_b >= 2:
            return {"unable_to_borrow": True }
        return {"unable_to_borrow": False }
    else:
        return{"unable_to_borrow": False}

        

#printing format for registered books   
#Inventory list
def inventory(bid:str):
    b =book[bid]
    return(
        f"\nName: {bid}\n"
        f"Tag: {b['tag']}\n"
        f"Author: {b['author']}\n"
        f"Date of Entry: {b['doe']}\n"
        f"Copies: {b['copies']}\n"  
        f"Genre: {b['genre']}"
)
 #printing format for registered students   
def student_registered(sid:str):
    s =student[sid]
    return(
        f"\nID: {sid}\n"
        f"Name: {s['name']}\n"
        f"Department: {s['department']}"
)

#create

def register_book(bid: str, tag:str, author:str, doe: str, copies: int, genre: str):
    if book_exists(bid):
        print(f'Book with ID already exists')
        return
    book[bid] = {
        'tag': tag,
        'author': author,
        'doe': doe,
        'copies': copies,
        'genre': genre
    }
    print(f"New book with name: {bid} has been added successfully.")
    
def register_student(sid: str, name: str, department: str):
    if student_exists(sid):
        print(f'Student with ID already exists')
        return
    student[sid] = {
        'name': name,
        'department': department,
        'borrowing_details': {
            'borrowed_book': [],
            'borrow_date': [],
            'return_date': [],
            'borrowing_history': []
        }
    }
    print(f"New student with ID: {sid} has been created successfully.")

    
#read

def read_book(bid: str):
    if book_exists(bid):
        print(inventory(bid))
    else:
        print(f"This book name: {bid} does not exist")
        
#List

def list_books():
    if not book:
        print("No books available in the store")
        return
    print("The books available include:")
    for bid in book:
        print(inventory(bid))
        
def list_students():
    if not student:
        print("No Students registered.")
        return
    print("The Students registered include:")
    for sid in student:
        print(student_registered(sid))
        
        
        
#borrow and checkout
#borrow
def borrow():
    print("You are now in the borrowing section")
    sid = input("Input your ID: ")
    if sid not in student:
        print(f"Student ID: {sid} not found")
        return
    check_history = input("Have you borrowed any books before?").strip().lower()
    borrowed_info = borrowed_book_checker(sid, check_history)
    if borrowed_info ["unable_to_borrow"]:
        print("You are ineligible to borrow, as you have pending borrrowed books beyond the threshold.")
        return
    else:
        list_books()
        book_choice = input("Which book do you want to borrow?").strip().lower()
        requested_copies = int(input(f"How many copies of {book_choice} do you want to borrow?").strip())
        if requested_copies > 2:
            print("You cannot borrow more than 2 books at a time!")
            return
        if book_exists(book_choice) and book[book_choice]['copies'] >=  requested_copies:
            proceed= input(f"{requested_copies} copy out of {book[book_choice]['copies']} of the book is available. Do you want to proceed to checkout?(yes/no)").strip().lower()
 #checkout
            if proceed == "yes":
                book[book_choice]['copies'] -= requested_copies
                current_date = datetime.today()  # gets today's date and time
                due_date = current_date + timedelta(days=7)
                current_date_str = current_date.strftime("%Y-%m-%d")
                due_date_str = due_date.strftime("%Y-%m-%d")
                student[sid]["borrowing_details"]["borrowed_book"].append(book_choice)
                student[sid]["borrowing_details"]["borrow_date"].append(current_date)
                student[sid]["borrowing_details"]["return_date"].append(due_date)
                student[sid]["borrowing_details"]["borrowing_history"].append(book_choice)
                print("Checkout Successful. Enjoy your books.")
                print(f"You borrrowed the book {current_date_str} and are to return it by {due_date_str}, or risk fines.")     
            else:
                print("Checkout cancelled.")
                return
        else:
            print(f"{book_choice} is not available or is not in stock.")
            return

#overdue
def overdue(sid: str, book_name: str):
    current_date = datetime.today().date()
    student_info = student[sid]["borrowing_details"]
    if book_name in student_info["borrowed_book"]:
        idx = student_info["borrowed_book"].index(book_name)
        return_date = student_info["return_date"][idx].date()
        if current_date > return_date:
            overdue_days = (current_date -  return_date).days
            fine = float(overdue_days * 150)
            return{"is_overdue": True, "days": overdue_days, "fine": fine}
        else:
            return{"is_overdue": False, "days": 0, "fine": 0.00}
    return{"is_overdue": False, "days": 0, "fine": 0.0}
    
   
#return and fines
def retun():
    sid= input("Enter your ID: ").strip().upper()
    if sid not in student:
        print("ID not found.")
        return
    return_option = input("What book do you want to return?").strip().lower()
    retun_quantity = int(input("How many copies do you want to return?"))
    if return_option in student[sid]['borrowing_details']['borrowed_book']:
        overdue_info= overdue(sid, return_option)
        if overdue_info["is_overdue"]:
            print(f"This book is overdue by {overdue_info['days']} day(s), and has incurred a fine of {overdue_info['fine']}") 
        else:
            print(f"{retun_quantity} copy(ies) of the book was returned on time. No fine")
            book[return_option]['copies'] += retun_quantity
            idx = student[sid]["borrowing_details"]["borrowed_book"].index(return_option)
            student[sid]["borrowing_details"]["borrowed_book"].pop(idx)
            student[sid]["borrowing_details"]["borrow_date"].pop(idx)
            student[sid]["borrowing_details"]["return_date"].pop(idx)
            print(f"{return_option} has been returned successfully.")
    else:  
        print("This book is not borrowed by you")
        return
    

#search
def search():
    search_option = input("What book are you looking for?").strip().lower()
    if book_exists(search_option) and book[search_option]['copies'] >= 1:
        print("The book is available.")
        choice_borrow = input("Do you want to borrow it?").lower()
        if choice_borrow == "yes":
            borrow()
        else:
            print("Have a good day")
    else:
        print("The book is not available or out of stock.")
        
        
#most popular
def most_popular(n=3):
    all_borrowed = []
    for sid, info in student.items():
        all_borrowed.extend(info["borrowing_details"]["borrowing_history"])
    
    book_counts = Counter(all_borrowed)
    top_borrowed = [title for title, _ in book_counts.most_common(n)]    
    if not top_borrowed:
        print("There are no popular books")
    else:
        print(f"The top {n} most popular books include: ")
        for i, book in enumerate(top_borrowed, start=1):
            print(f"{i}.{book} ")

def delete_books():
    delete_choice = input("What book do you want to remove?").strip().lower()
    reask = input("Are you sure you want to delete the book?").strip().lower()
    if book_exists(delete_choice) and reask == "yes":
        del book[delete_choice]
    else:
        print("This book does not exist/Accidental delete attempt")
            
#password
def password() ->bool: 
    pass_code = input("Please type in the password?").strip()
    if pass_code == "Cisco321$":
        print("Access Granted")
        return True
    else:
        print("Access Denied. Incorrect password.")
        return False 


#main code
def main():
    while True:
        print("=======Library Management System=======")
        identification = input("Welcome to Library Management system. Are you a student or staff?").strip().lower()
        if identification == "student":
            print("Student segment")
            print("1) Create Account")
            print("2) Search Books")
            print("3) All books")
            print("4) Borrow Books")
            print("5) Return Books")
            print("0) Exit")
        elif identification == "staff":
            print("Librarian Segment")
            print("a) Add Books")
            print("b) Delete Books")
            print("c) Most Popular")
            print("d) Registered Students")
            print("0) Exit")
        else:
            print("This system is not for you. Goodbye!")
            return
            
        choice = input("What option are you interested in?").strip().lower()
        
        if choice == "1":
            sid = input("What is your ID: ").strip()
            name = input("What is your full name: ").strip().lower()
            department = input("What is your department: ").strip().lower()
            register_student(sid, name, department)
        elif choice == "2":
            search()
            research = input("Do you want to search for another book?").strip().lower()
            if research == "yes":
                continue
        elif choice == "3":
            list_books()
        elif choice == "4":
            borrow()
        elif choice == "5":
            retun()
        elif choice == "a":
            if password():
                bid = input("Input book name: ").strip().lower()
                tag = input("Input book unique tag: ").strip()
                author = input("Who wrote the book: ").strip().lower()
                doe = input("Input date of entry(dd-mm-yy): ").strip().lower()
                copies = int(input("How many copies are you bringing in: ").strip())
                genre = input("What genre is the book(science, fiction, engineering):").strip().lower()
                register_book(bid, tag, author, doe, copies, genre)
        elif choice == "b":
            if password():
                delete_books()
        elif choice == "c":
            if password():
                most_popular()
        elif choice == "d":
            if password():
                list_students()
        elif choice == "0":
            print("Thanks for checking in. Readers are leaders!")
            break
        else:
            print("Wrong input, Please try again.")
            
if __name__  == "__main__":
    main()
```
